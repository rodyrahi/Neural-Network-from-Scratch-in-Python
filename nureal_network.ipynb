```python
import numpy as np
from nnfs.datasets import spiral_data

np.random.seed(0)

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

# Generate dataset
X, y = spiral_data(100, 3)  # X shape (300, 2), y shape (300,)

# Hyperparameters
learning_rate = 0.001
neurons = 4

# Initialize weights and biases
weights = np.random.randn(2, neurons)  # Shape (2, 4) since X has 2 features
biases = np.zeros((1, neurons))  # Shape (1, 4)

for i in range(100):
    # Forward Pass
    z = np.dot(X, weights) + biases  # Shape (300, 4)
    a = relu(z)

    # Loss Calculation (Mean Squared Error)
    loss = np.mean(a ** 2)

    # Backward Pass
    drelu = relu_derivative(z)
    dz = a * drelu  # Element-wise multiplication

    dweights = np.dot(X.T, dz) / X.shape[0]  # Normalize gradient by batch size
    dbiases = np.sum(dz, axis=0, keepdims=True) / X.shape[0]

    # Update weights and biases
    weights -= learning_rate * dweights
    biases -= learning_rate * dbiases

    if (i + 1) % 10 == 0:
        print(f"Iteration {i+1}: Loss: {loss:.6f}")

```

```python
import numpy as np


# Activation functions
def relu(x):
    return np.maximum(0, x)


def relu_derivative(x):
    return np.where(x > 0, 1, 0)


def softmax(x):
    exp_values = np.exp(x - np.max(x, axis=1, keepdims=True))  # Prevent overflow
    return exp_values / np.sum(exp_values, axis=1, keepdims=True)


# Spiral dataset generator
def spiral_data(samples, classes):
    np.random.seed(0)
    X = np.zeros((samples * classes, 2))
    y = np.zeros(samples * classes, dtype=int)
    for class_number in range(classes):
        ix = range(samples * class_number, samples * (class_number + 1))
        r = np.linspace(0.0, 1, samples)
        t = np.linspace(class_number * 4, (class_number + 1) * 4, samples) + np.random.randn(samples) * 0.2
        X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]
        y[ix] = class_number
    return X, y


# Fully connected layer with He initialization
class LayerDense:
    def __init__(self, n_inputs, n_neurons, learning_rate=0.01):
        self.weights = np.random.randn(n_inputs, n_neurons) * np.sqrt(2 / n_inputs)  # He initialization
        self.biases = np.zeros((1, n_neurons))
        self.learning_rate = learning_rate

    def forward(self, inputs):
        self.inputs = inputs
        self.output = np.dot(inputs, self.weights) + self.biases

    def backward(self, dvalues):
        self.dweights = np.dot(self.inputs.T, dvalues)
        self.dbiases = np.sum(dvalues, axis=0, keepdims=True)
        self.dinputs = np.dot(dvalues, self.weights.T)

        self.weights -= self.learning_rate * self.dweights
        self.biases -= self.learning_rate * self.dbiases


class LossCategoricalCrossEntropy:
    def forward(self, y_pred, y_true):
        samples = len(y_pred)
        y_pred_clipped = np.clip(y_pred, 1e-7, 1 - 1e-7)  # Prevent log(0)

        if len(y_true.shape) == 1:  # If y_true is categorical labels
            correct_confidences = y_pred_clipped[range(samples), y_true]
        else:  # If y_true is one-hot encoded
            correct_confidences = np.sum(y_pred_clipped * y_true, axis=1)

        return -np.mean(np.log(correct_confidences))

    def backward(self, dvalues, y_true):
        samples = len(dvalues)
        
        # Convert labels to one-hot if they are categorical
        if len(y_true.shape) == 1:
            y_true = np.eye(dvalues.shape[1])[y_true]  

        # Softmax + Cross-Entropy Derivative
        self.dinputs = (dvalues - y_true) / samples



# Creating dataset
X, y = spiral_data(100, 3)


layer1 = LayerDense(2, 4, learning_rate=0.01)
layer2 = LayerDense(4, 4, learning_rate=0.01)
layer3 = LayerDense(4, 3, learning_rate=0.01)

loss_function = LossCategoricalCrossEntropy()


def plot_decision_boundary(model, X, y):
    # Define the grid
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), 
                         np.linspace(y_min, y_max, 100))

    # Create a dataset from the grid and pass it through the model
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    layer1.forward(grid_points)
    layer1.output = relu(layer1.output)

    layer2.forward(layer1.output)
    layer2.output = relu(layer2.output)

    layer3.forward(layer2.output)
    predictions = np.argmax(softmax(layer3.output), axis=1)
    predictions = predictions.reshape(xx.shape)

    # Plot decision boundary
    plt.contourf(xx, yy, predictions, alpha=0.3, cmap=plt.cm.Spectral)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral, edgecolors="k")
    plt.title("Classification Decision Boundary")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

# Train the model
epochs = 100000
for epoch in range(epochs):
    # Forward pass
    layer1.forward(X)
    layer1.output = relu(layer1.output)

    layer2.forward(layer1.output)
    layer2.output = relu(layer2.output)

    layer3.forward(layer2.output)
    layer3.output = softmax(layer3.output)

    # Loss calculation
    loss = loss_function.forward(layer3.output, y)

    # Accuracy
    predictions = np.argmax(layer3.output, axis=1)
    accuracy = np.mean(predictions == y)

    # Backward pass
    loss_function.backward(layer3.output, y)
    layer3.backward(loss_function.dinputs)

    dvalues = relu_derivative(layer2.output) * layer3.dinputs
    layer2.backward(dvalues)

    dvalues = relu_derivative(layer1.output) * layer2.dinputs
    layer1.backward(dvalues)

    if epoch % 50 == 0:
        print(f'Epoch {epoch}, Loss: {loss:.4f}, Accuracy: {accuracy:.4f}')

# Plot classification boundary
plot_decision_boundary(layer3, X, y)
```
