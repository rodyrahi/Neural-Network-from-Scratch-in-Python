```python
def relu(x):
    return np.maximum(0 ,x)


def relu_derivative(x):
    return np.where(x > 0, 1, 0)
```

```python
import numpy as np

np.random.seed(0)



learning_rate = 0.001

inputs = np.array([[1, 2, 3, 4]])  # Reshaped to (1, 4)

neurons = 4

weights = np.random.randn(4, neurons)  # Shape (4,4)
biases = np.zeros((1, neurons))  # Shape (1, 4)

for i in range(100):
    # Forward Pass
    z = np.dot(inputs, weights) + biases  # Shape (1, 4)
    a = relu(z)

    # Backward Pass
    drelu = relu_derivative(z)
    
    dz = a * drelu  # Element-wise multiplication (Hadamard product)
    dweights = np.dot(inputs.T, dz)  # Shape (4,4)
    dbiases = np.sum(dz, axis=0, keepdims=True)  # Shape (1,4)
    
    dinputs = np.dot(dz, weights.T)  # Shape (1,4)

    # Gradient Descent Update
    weights -= learning_rate * dweights
    biases -= learning_rate * dbiases

    print(f"Iteration {i+1}: Weights:\n{weights}\nBiases:\n{biases}")

    y = np.sum(a)

    # Calculate loss
    loss = y ** 2

    print(loss)





```
